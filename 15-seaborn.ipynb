```python
sns.set(style="darkgrid")
t = sns.load_dataset("tips")

Male = t.query("sex == 'Male'")
Female = t.query("sex == 'Female'")

# 設定視窗大小
f, ax = plt.subplots(figsize=(30, 10))
ax.set_aspect("equal")

# 繪製核密度圖
ax = sns.kdeplot(Male.total_bill, Male.tip,
                 cmap="Blues", shade=True, shade_lowest=False)
ax = sns.kdeplot(Female.total_bill, Female.tip,
                 cmap="Reds", shade=True, shade_lowest=False)

# 加上標籤 
red = sns.color_palette("Blues")[-2]
blue = sns.color_palette("Reds")[-2]
ax.text(5, 8.2, "Male", size=16, color=blue)
ax.text(30, 8, "Female", size=16, color=red)
```

### planets 行星資料集包含軌道週期, 質量, 距離 以及年分。

```python
sns.set(style="darkgrid")
iris = sns.load_dataset("iris")

setosa = iris.query("species == 'setosa'")
#setosa = iris.loc[iris.species == "setosa"]
virginica = iris.query("species == 'virginica'")
#virginica = iris.loc[iris.species == "virginica"]

f, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

ax = sns.kdeplot(setosa.sepal_width, setosa.sepal_length,
                 cmap="Reds", shade=True, shade_lowest=False)
ax = sns.kdeplot(virginica.sepal_width, virginica.sepal_length,
                 cmap="Blues", shade=True, shade_lowest=False)

red = sns.color_palette("Reds")[-2]
blue = sns.color_palette("Blues")[-2]
ax.text(2.5, 8.2, "virginica", size=16, color=blue)
ax.text(3.8, 4.5, "setosa", size=16, color=red)
```
